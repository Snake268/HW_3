```python
import pandas as pd

# Загрузка данных
df = pd.read_csv('train.csv')

# Основная информация о датасете
df.info()
df.describe()
df.isnull().sum()

# Процент выживаемости у каждого класса пассажиров
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print(survival_rate_by_class)

# Извлечение имен
df['FirstName'] = df['Name'].str.extract(r'([A-Za-z]+)\.', expand=False)

# Самое популярное мужское и женское имя
male_names = df[df['Sex'] == 'male']['FirstName'].value_counts()
most_popular_male_name = male_names.idxmax()

female_names = df[df['Sex'] == 'female']['FirstName'].value_counts()
most_popular_female_name = female_names.idxmax()

print(f"Самое популярное мужское имя: {most_popular_male_name}")
print(f"Самое популярное женское имя: {most_popular_female_name}")

# Самое популярное имя в каждом классе
popular_names_by_class = df.groupby(['Pclass', 'Sex'])['FirstName'].agg(lambda x: x.value_counts().idxmax())
print(popular_names_by_class)

# Пассажиры старше 44 лет
passengers_older_than_44 = df[df['Age'] > 44]
print(passengers_older_than_44)

# Молодые мужчины младше 44 лет
young_male_passengers = df[(df['Age'] < 44) & (df['Sex'] == 'male')]
print(young_male_passengers)

# Количество n-местных кабин
cabin_counts = df['Cabin'].value_counts()
n_occupancy_counts = cabin_counts[cabin_counts >= 2]
print(n_occupancy_counts)
```
